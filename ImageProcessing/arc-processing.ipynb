```python
'''
根据label[2013]文件裁剪label[1999]tif文件
'''



import os
import rasterio
from rasterio.mask import mask
from rasterio.windows import from_bounds
from shapely.geometry import box
import numpy as np

# 输入与输出文件夹路径
input_folder = r"H:\ArcGIS File\extracting-slums-from-satellite-imagery\5_training_data_rio_rio2019&2024_cell0.5_tile2024_stride0\labels[2013to2019]"
output_folder = r"H:\ArcGIS File\extracting-slums-from-satellite-imagery\5_training_data_rio_rio2019&2024_cell0.5_tile2024_stride0\labels[1999]"
target_tif_file = r"C:\Users\YANGRUYI\Documents\ArcGIS\Projects\MyProject\Limite_Favelas_1999_raste.tif"

# 创建输出文件夹
os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.endswith(".tif"):
        boundary_tif_path = os.path.join(input_folder, filename)

        # 读取裁剪边界的 TIFF 文件
        with rasterio.open(boundary_tif_path) as boundary_dataset:
            boundary_bounds = boundary_dataset.bounds
            boundary_geom = [box(*boundary_bounds)]

        # 读取被裁剪的 TIFF 文件
        with rasterio.open(target_tif_file) as target_dataset:
            target_bounds = target_dataset.bounds
            
            # 检查裁剪窗口是否超出目标范围
            if (
                boundary_bounds.left < target_bounds.left or
                boundary_bounds.right > target_bounds.right or
                boundary_bounds.bottom < target_bounds.bottom or
                boundary_bounds.top > target_bounds.top
            ):
                print(f"裁剪窗口超出了目标 TIFF 的范围: {filename}")
                
                # 创建覆盖裁剪窗口的完整窗口
                full_window = from_bounds(
                    boundary_bounds.left, boundary_bounds.bottom,
                    boundary_bounds.right, boundary_bounds.top,
                    target_dataset.transform
                )
                
                # 使用 boundless=True 和 fill_value=0 填充超出范围的部分
                out_image = target_dataset.read(
                    window=full_window,
                    boundless=True,
                    fill_value=0
                )

                # 计算输出栅格的 transform
                out_transform = rasterio.windows.transform(full_window, target_dataset.transform)
            else:
                # 裁剪窗口完全在目标范围内，直接裁剪
                out_image, out_transform = mask(target_dataset, boundary_geom, crop=True, nodata=0)

            # 更新元数据
            out_meta = target_dataset.meta.copy()
            out_meta.update({
                "driver": "GTiff",
                "height": out_image.shape[1],
                "width": out_image.shape[2],
                "transform": out_transform,
                "crs": target_dataset.crs,
                "nodata": 0
            })

            # 保存裁剪后的 TIFF 文件
            out_tif_path = os.path.join(output_folder, filename)
            with rasterio.open(out_tif_path, "w", **out_meta) as dest:
                dest.write(out_image)
            # print("裁剪完成，裁剪后的文件保存到:", out_tif_path)

print("批量裁剪完成，裁剪后的文件保存到:", output_folder)
```

```python
import os
from osgeo import gdal, gdal_array
import numpy as np
# from scipy.ndimage import zoom
from PIL import Image

# 输入和输出文件夹路径
input_folder = r'H:\ArcGIS File\extracting-slums-from-satellite-imagery\5_training_data_rio_rio2019&2024_cell0.5_tile2024_stride0\labels[1999]_raw'
output_folder = r'H:\ArcGIS File\extracting-slums-from-satellite-imagery\5_training_data_rio_rio2019&2024_cell0.5_tile2024_stride0\labels[1999]'

# 遍历输入文件夹中的所有 TIFF 文件
for filename in os.listdir(input_folder):
    if filename.endswith('.tif'):
        input_tif = os.path.join(input_folder, filename)
        output_tif = os.path.join(output_folder, filename)
        
        # 打开输入 TIFF 文件
        dataset = gdal.Open(input_tif)
        if dataset is None:
            print("无法打开输入文件:", input_tif)
        else:
            # 读取第 4 波段数据
            band_4 = dataset.GetRasterBand(4)
            band_4_data = band_4.ReadAsArray()

            # 将值 255 转换为 1，其它值为 0（如需要二值化）
            band_4_data_binary = np.where(band_4_data == 255, 1, 0)

            # 使用 Pillow 调整分辨率到 1024x1024（或其他目标尺寸）
            image = Image.fromarray(band_4_data_binary.astype(np.uint8))
            image_resized = image.resize((1024, 1024), Image.Resampling.NEAREST)
            band_4_data_resized = np.array(image_resized)

            # 创建输出单波段 TIFF 文件
            driver = gdal.GetDriverByName('GTiff')
            out_dataset = driver.Create(output_tif, 1024, 1024, 1, gdal.GDT_Byte)

            # 设置地理信息
            geo_transform = dataset.GetGeoTransform()
            projection = dataset.GetProjection()
            if geo_transform:
                out_dataset.SetGeoTransform(geo_transform)
            if projection:
                out_dataset.SetProjection(projection)

            # 写入第 4 波段数据
            out_band = out_dataset.GetRasterBand(1)
            out_band.WriteArray(band_4_data_resized)
            out_band.FlushCache()

            # 显式关闭文件
            out_dataset = None
            dataset = None

            print(f"处理后的文件（仅第 4 波段）已保存到: {output_tif}")
```

```python
# Step 1: 裁剪矢量文件
with arcpy.da.SearchCursor(fishnet, ["FID", "SHAPE@"]) as cursor:
    for fid, geometry in cursor:
        # 定义裁剪后的矢量文件路径
        clipped_vector = os.path.join(output_folder, f"tile_{fid:04d}.shp")
        
        # 裁剪矢量数据
        arcpy.analysis.Clip(
            in_features=input_vector,
            clip_features=geometry,
            out_feature_class=clipped_vector
        )
        print(f"裁剪完成：tile_{fid:04d}.shp")
        
        # Step 2: 栅格化裁剪后的矢量文件
        rasterized_vector = os.path.join(output_folder, f"tile_{fid:04d}.tif")
        arcpy.conversion.PolygonToRaster(
            in_features=clipped_vector,
            value_field="code",  
            out_rasterdataset=rasterized_vector,
            cell_assignment="MAXIMUM_AREA",
            priority_field="NONE",
            cellsize=cell_size
        )
        print(f"栅格化完成：tile_{fid:04d}.tif")

print("所有裁剪和栅格化操作完成！")
```

### method 2  基于多边形顶点的预测

```python
import geopandas as gpd
import pandas as pd

def extract_coordinates(row):
    """
    提取多边形的边界坐标。
    """
    if row.geometry.geom_type == 'Polygon':
        return list(row.geometry.exterior.coords)
    elif row.geometry.geom_type == 'MultiPolygon':
        print('{} is MultiPolygon'.format(row['cod_favela']))
        # 如果是 MultiPolygon，提取每个多边形的外部边界
        return [list(poly.exterior.coords) for poly in row.geometry.geoms]
    else:
        raise ValueError(f"Unsupported geometry type: {row.geometry.geom_type}")

def process_shapefile(shapefile_path):
    """
    处理单个 Shapefile：提取 cod_favela 和边界坐标。
    """
    gdf = gpd.read_file(shapefile_path)
    gdf['coordinates'] = gdf.apply(extract_coordinates, axis=1)
    return gdf[['cod_favela', 'coordinates']]  # 仅保留 ID 和坐标列

def merge_shapefiles(shapefiles):
    """
    合并多个 Shapefile，保留共有的 cod_favela。
    """
    processed = []
    for shp in shapefiles:
        gdf = process_shapefile(shp)
        year = os.path.basename(shp).split('_')[-1].split('.')[0]  # 提取年份
        gdf = gdf.rename(columns={'coordinates': f'coordinates_{year}'})
        processed.append(gdf)
    
    merged = processed[0]
    for df in processed[1:]:
        merged = merged.merge(df, on='cod_favela', suffixes=('', '_other'))
    return merged

# 示例 Shapefile 路径
shapefiles = [r"H:\ArcGIS File\favela boundary\Limite_Favelas_1999\Limite_Favelas_1999.shp",
               r"H:\ArcGIS File\favela boundary\Limite_Favelas_2008\Limite_Favelas_2008.shp", 
               r"H:\ArcGIS File\favela boundary\Limite_Favelas_2013\Limite_Favelas_2013.shp",
               r'H:\ArcGIS File\favela boundary\Limite_Favelas_2019\Limite_Favelas_2019.shp'
               ]

# 处理并合并 Shapefile
merged_df = merge_shapefiles(shapefiles)

# 保存为 CSV
# merged_df.to_csv("output_directory/merged_favela.csv", index=False)

# print("合并后的数据已保存为 merged_favela.csv")
```
